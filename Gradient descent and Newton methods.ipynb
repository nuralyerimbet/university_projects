# Unconstrained minimization

## Gradinet descent method:

### 1 problem: ð‘“(ð‘¥)=sinx+x^4

```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
epsilon = 0.001 
alpha = 1/2 
betta = 1/2
x = 2 # starting point
```

```python
def func(x):
    y = np.sin(x) + x**4
    return y
def grad(x):
    y_prime = np.cos(x) + 4*x**3
    return y_prime
def hessian(x):
    y_prime_prime = -np.sin(x) + 12*x**2
    return y_prime_prime
```

```python
list_of_x = [x,]
dx = -grad(x) # Step

while abs(grad(x)) > epsilon:
    t = 1
    while func(x + t*dx) > func(x) + alpha*t*grad(x)*dx: # Backtracking line search
        t *= betta  
    x = x + t*dx
    
    list_of_x.append(x)
print(f"{len(list_of_x)} points: ", *list_of_x)
```

### Graph

```python
list_of_y = []
for i in list_of_x:
    y = func(i)
    list_of_y.append(y)
    
X = np.linspace(-2,2,101)
Y = np.zeros(101)
for i in range(len(X)):
    Y[i] = func(X[i])
    
fig, axes = plt.subplots(figsize=(8, 8))
axes.plot(list_of_x, list_of_y, 'r-o')
axes.plot(X, Y, 'b-')
axes.grid()
axes.set_title('1st function')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
#axes.legend()
```

## Newton's method:

```python
list_of_x2 = [x,]
dx = -grad(x) # Step

while abs(grad(x)**2/hessian(x))/2 > epsilon:
    t = 1
    while func(x + t*dx) > func(x) + alpha*t*grad(x)*dx: # Backtracking line search
        t *= betta  
    x = x + t*dx
    
    list_of_x2.append(x)
print(f"{list_of_x2} poits: ", *list_of_x2)
```

### Graph

```python
list_of_y2 = []
for i in list_of_x2:
    y = func(i)
    list_of_y2.append(y)
    
X = np.linspace(-2,2,101)
Y = np.zeros(101)
for i in range(len(X)):
    Y[i] = func(X[i])
    
fig, axes = plt.subplots(figsize=(8, 8))
axes.plot(list_of_x2, list_of_y2, 'r-o')
axes.plot(X, Y, 'b-')
axes.grid()
axes.set_title('1st function')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
#axes.legend()
```

# Constrained minimization

## Gradient descent method:

## 2nd problem: ð‘“(ð‘¥)=2ð‘¥log(2ð‘¥+3)âˆ’3(log1/(2ð‘¥+3))

```python
def func(x):
    y = (2*x+3)*math.log(2*x+3,10)
    return y
def grad(x):
    y_prime = 2*math.log(2*x+3,10) + 2/math.log(10)
    return y_prime
def hessian(x):
    y_prime_prime = 2/((2*x+3)*math.log(10))
    return y_prime_prime
```

```python
list_of_x = [x,]
dx = -grad(x) # step

while abs(grad(x)) > epsilon:
    t = 1
    while True:
        if x + t*dx > -3/2:
            while func(x + t*dx) > func(x) + alpha*t*grad(x)*dx: # Backtraking line search
                t *= betta  
            x = x + t*dx
            break
        t *= betta
    
    list_of_x.append(x)
print(f"{list_of_x} poits: ", *list_of_x)
```

### Graph

```python
list_of_y = []
for i in list_of_x:
    y = func(i)
    list_of_y.append(y)
    
X = np.linspace(-2.9999/2,2,101)
Y = np.zeros(101)
for i in range(len(X)):
    Y[i] = func(X[i])
    
fig, axes = plt.subplots(figsize=(8, 8))
axes.plot(list_of_x, list_of_y, 'r-o')
axes.plot(X, Y, 'b-')
axes.grid()
axes.set_title('2nd function')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
#axes.legend()
```

## Newton's method:

```python
list_of_x2 = [x,]
dx = -grad(x)/hessian(x) # step

while abs(grad(x)**2/hessian(x))/2 > epsilon:
    t = 1
    while True:
        if x + t*dx > -3/2:
            while func(x + t*dx) > func(x) + alpha*t*grad(x)*dx:
                t *= betta
            x = x + t*dx
            break
        t *= betta
    
    list_of_x2.append(x)
print(f"{list_of_x2} points:", *list_of_x2)
```

### Graph

```python
list_of_y = []
for i in list_of_x2:
    y = func(i)
    list_of_y.append(y)
    
X = np.linspace(-2.9999/2,2,101)
Y = np.zeros(101)
for i in range(len(X)):
    Y[i] = func(X[i])
    
fig, axes = plt.subplots(figsize=(8, 8))
axes.plot(list_of_x2, list_of_y, 'r-o')
axes.plot(X, Y, 'b-')
axes.grid()
axes.set_title('2nd function')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
#axes.legend()
```
